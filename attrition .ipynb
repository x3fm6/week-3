```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import normalize 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier #Ik stop ze er gewoon allemaal meteen in
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

df = pd.read_csv("attrition.csv")
df
```

#Ik heb gekozen voor de employee attritition data set omdat hij mij het meest interessant leek

#de rest lijken alsof ze oninteressant zullen zijn in vergelijking met employee attritition 

#en "predict attrition" klinkt cooler want ik het doet me denken aan attrition in een oorlog of zoiets, ook al gaat het maar over medewerkers

```python
dummies = pd.get_dummies(df['Attrition'])
df = pd.concat([df, dummies], axis=1) 
df.head()
```

#Ik zorg er meteen voor dat attrition numeriek kan worden weergegeven 

```python
X = df[['YearsAtCompany', 'TotalWorkingYears', 'DistanceFromHome', 'Education', 'TrainingTimesLastYear', 'WorkLifeBalance']]
y = df['Yes']
X
```

#Ik kies deze 7 variabelen voor de volgende redenen
#Yes: Het is interessanter om te weten welke mensen wel attritrion hebben
#YearsAtCompany: hoe lang ze bij het bedrijf werken kan misschien die attritrion veroorzaken
#TotalWorkingYears: hetzelfde als hierboven
#DistanceFromHome: hoe lang iemand moet reizen is altijd heel belangrijk
#Education: heeft hun niveau van educatie misschien invloed
#TrainingTimesLastYear: Ik wil zien of meer training dingen verbeteren of niet
#WorklifeBalance: Te veel werken is heel vaak een rede voor burn out 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
X_train.head() #ik split de training en de test
```

```python
knn = KNeighborsClassifier(5) 
knn = knn.fit(X_train, y_train) 
knn
```

```python
knn.score(X_test, y_test)
```

#hier volg ik eigenlijk gewoon het proces dat we eerder hebben gevolgd zodat ik de knn score kan zien

#Door het aantal mensen die wel attrition hebben als Y te nemen kan ik de invloed van die variabelen op het aantal attritrion berekenen 

#Ik zie hier dus dat het ongeveer 80% accuraat is als het gaat om berekenen van attrition 

```python
y_test_pred = knn.predict(X_test) 
cm = confusion_matrix(y_test, y_test_pred) 
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Yes attrition (actual)', 'no attrition (actual)'], columns = ['Yes attrition (predicted)', 'No attrition (predicted)']) 
conf_matrix
```
